```python
import pandas as pd
```

```python
df = pd.read_excel("food production.xlsx",sheet_name="Sheet1")
df.head()
```

```python
df['Year'] = df['Year'].apply(lambda x: int(str(x)[:4]))
```

```python
df.head(2)
```

```python
df.dtypes
```

```python
df['Year'] = pd.to_datetime(df['Year'],format='%Y')
```

```python
df = df.sort_values(by=['Year','State / Union Territory'])
```

```python
df.set_index('Year',inplace=True)
```

```python
df.head(3)
```

```python
df.to_excel("food production_cleaned.xlsx")
```

```python
df = pd.read_excel("food production_cleaned.xlsx")
```

```python
df.set_index("Year")
```

```python
df[df['State / Union Territory'] == "Andhra Pradesh"].plot(
    x="Year",
    y="Pulses",
    figsize=(10,6),
    title="Pulses Production in Andhra Pradesh"
)
```

```python
import matplotlib.pyplot as plt 
states = df['State / Union Territory'].unique()  # Get list of all unique states 
for state in states:

    state_df = df[df['State / Union Territory'] == state]

    plt.figure(figsize=(10, 6))

    plt.title(f"Pulses Production Over Years - {state}")

    plt.plot(state_df['Year'], state_df['Pulses'], marker='o')

    plt.xlabel('Year')

    plt.ylabel('Pulses Production')

    plt.grid(True)

    plt.show()
```

```python
import matplotlib.pyplot as plt

# Pivot Data: States become columns; Year is index
pivot_df = df.pivot(index="Year", columns="State / Union Territory", values="Pulses")

# Plot all states on one figure
plt.figure(figsize=(18, 10))  # Increase size for readability
for state in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[state], label=state)

plt.xlabel('Year')
plt.ylabel('Pulses Production')
plt.title('Pulses Production Over Years by State')
plt.legend(loc="upper left", bbox_to_anchor=(1, 1), ncol=1, frameon=False, fontsize='small')  # Legend outside plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
